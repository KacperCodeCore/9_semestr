```python
for i in range(20):
    x = i
    for y in range(20):
        multi = i * y
        if multi > 100:
            break
    print(f"x: {i}")
    print(f"multi: {multi}")
    print(f"x*multi: {x*multi}")
    answer = x * multi + y
    print(f"answer: {answer}")

    y = x - 1
    answer = x * multi + y
    print(x*multi)
    print(answer)
    print(multi + y)
```
